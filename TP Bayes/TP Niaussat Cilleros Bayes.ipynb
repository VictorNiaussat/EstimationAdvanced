```python
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 0.001
b= 0.001


x = np.linspace(gamma.ppf(0.001, a),
                gamma.ppf(0.999, a), 100)

plt.style.use('ggplot')

plt.plot(x, gamma.pdf(x, a,scale=1/b),
       'b-', lw=2, alpha=0.6, label='gamma pdf')
plt.title("$\Gamma$ Density for parameters $a=$" +"{:.1e}".format(a) +" et $b=$"+"{:.1e}".format(b) )
plt.savefig("gamma.png")
```
