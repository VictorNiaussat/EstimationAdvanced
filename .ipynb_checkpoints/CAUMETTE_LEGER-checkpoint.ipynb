```python
# Import librairies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform
```

```python
from scipy.stats import norm

def gm_pdf(x, mu, sigma, p):
    resultat = 0
    if len(mu) != len(p):
        print('Erreur de dimension sur la moyenne')
    elif len(sigma) != len(p):
            print('Erreur de dimension sur la variance')
    else:
        for i in range(0, len(p)):
            resultat = resultat + p[i] * norm.pdf(x = x, loc = mu[i], scale = sigma[i])
    return resultat
```

```python
dataframe = pd.read_excel('2223G3-SDIA-Galaxy.xlsx')

dataframe.head()
```

```python
mu = dataframe.mean()
sigma = dataframe.var()

print('mu : {}'.format(mu))
print('sigma : {}'.format(sigma))
```

```python
plt.plot(dataframe, 'x')
plt.xlabel('Planet index')
plt.ylabel('Velocity')
plt.show()
```

```python
plt.hist(dataframe, bins = 30, density = True, color = 'yellow', edgecolor = "red")
plt.title('Histogramme des vitesses des planètes')
plt.xlabel('Données')
plt.ylabel('Index')
plt.show()
```

```python
def algoEM(nbEchantillon, mu_em, sigma_em, alpha_em):
    # Initialisation
    nbMaxIterations = 40
    nbIteration = 1 #Initialisation de la variable d'arrêt
    nbComposante = len(alpha_em) #Nombre de composantes du mélange
    nbDonnees = len(dataframe)  #Nombre de données
    p = np.zeros(shape=(nbComposante, nbDonnees))
    #Déclaration et initialisation de la matrice qui va contenir les probabilités
    #p(k|x,theta_courant)
    
    alpha_em_new = alpha_em
    sigma_em_carre_new = sigma_em
    mu_em_new = mu_em
    donneesP = np.zeros(shape=(nbEchantillon))

    while nbIteration < nbMaxIterations:
        for n in range(0, nbDonnees, 1):
            for k in range(0, nbComposante, 1):
                p[k, n] = alpha_em[k] * norm.pdf(x = dataframe.iloc[n, 0], loc = mu_em[k], scale = sigma_em[k])

            p[:, n] = p[:, n] / np.sum(p[:, n])

        for k in range(0, nbComposante, 1):
            alpha_em_new[k] = np.sum(p[k,:]) / nbDonnees

            for n in range(0, nbDonnees, 1):
                donneesP[n] = dataframe.iloc[n, 0] * p[k, n]

            mu_em_new[k]  = np.sum(donneesP) / np.sum(p[k, :])

            for n in range(nbDonnees):
                donneesP[n] = ((dataframe.iloc[n, 0] - mu_em_new[k]) ** 2) * p[k, n]

            sigma_em_carre_new[k] = np.sum(donneesP) / np.sum(p[k, :])

        mu_em = mu_em_new
        sigma_em = np.sqrt(sigma_em_carre_new)
        alpha_em = alpha_em_new
        nbIteration = nbIteration + 1
        
    return mu_em, sigma_em, alpha_em

def plot(mu_em, sigma_em, alpha_em):
    plt.hist(dataframe, bins = 30, density = True, color = 'yellow', edgecolor = "red")

    x = np.arange(8000, 35000, 100)
    pointPdfEstime = gm_pdf(x, mu_em, sigma_em, alpha_em)
    plt.plot(x, pointPdfEstime, 'b-', label = 'Estimée')
    plt.legend(loc='upper left', shadow=True, fontsize='x-large')

    plt.xlabel('Vitesse')
    plt.ylabel('Probabilité')

    plt.show()
```

### Simulation d'un mélange de 4 gaussiennes

```python
nbEchantillon = 100
mu = np.array([10000, 20000, 22500, 33000])
sigma = np.array([1000, 1000, 1000, 1000])
alpha = np.array([0.15, 0.35, 0.35, 0.15])
```

```python
mu_em, sigma_em, alpha_em = algoEM(nbEchantillon, mu, sigma, alpha)
```

```python
print('Les paramètres estimés sont : ')
print('Moyennes des composantes du mélange', mu_em)
print('Ecrat type des composantes du mélange', sigma_em)
print('Probabilités des composantes du mélange', alpha_em)
```

```python
print('La somme des probabilités des composantes du mélange vaut : ', np.sum(alpha_em))
```

```python
plot(mu_em, sigma_em, alpha_em)
```

### Simulation d'un mélange de 6 gaussiennes

```python
nbEchantillon = 100
mu = np.array([10000, 16000, 19500, 22500, 27000, 33000])
sigma = np.array([1000, 500, 1500, 1500, 500, 1000])
alpha = np.array([0.12, 0.03, 0.35, 0.35, 0.03, 0.12])

mu_em, sigma_em, alpha_em = algoEM(nbEchantillon, mu, sigma, alpha)
```

```python
print('La somme des probabilités des composantes du mélange vaut : ', np.sum(alpha_em))
```

```python
plot(mu_em, sigma_em, alpha_em)
```

### Implémentation avec sklearn

```python
from sklearn.mixture import GaussianMixture
```

```python
X = dataframe
gm = GaussianMixture(n_components=4, random_state=0).fit(X)
mu_em, sigma_em, alpha_sklearn = gm.means_, gm.covariances_, gm.weights_
```

```python
mu_sklearn = [x[0] for x in mu_em]
sigma_sklearn = [x[0][0] for x in sigma_em]

print('Les paramètres estimés sont : ')
print('Moyennes des composantes du mélange : \n{}\n'.format(mu_sklearn))
print('Ecrat type des composantes du mélange : \n{}\n'.format(sigma_sklearn))
print('Probabilités des composantes du mélange : \n{}'.format(alpha_sklearn))
```

```python
plot(mu_sklearn, sigma_sklearn, alpha_sklearn)
```
